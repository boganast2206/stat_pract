## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
print("Введите матрицу (пустая строка для окончания ввода): ")
matrix = []
while True:
    row_input = input()
    if not row_input:
        break
    row = list(map(int, row_input.split()))
    matrix.append(row)

direction = input("Введите 'left' или 'right': ")

if direction == 'right':
    rotated = [list(row)[::-1] for row in zip(*matrix)]
elif direction == 'left':
    rotated = [list(row) for row in zip(*matrix)][::-1]

for row in rotated:
    print(row)
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
from datetime import datetime
from typing import List, Dict, Optional, Union


class Book:
    def __init__(self, title: str, author: str, year: int, isbn: str, copies: int = 1):

        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies
        print(f"Книга '{title}' добавлена в библиотеку. Экземпляров: {copies}")

    def get_info(self) -> str:
        return f"'{self.title}' | {self.author} | {self.year} г. | ISBN: {self.isbn} | Экземпляров: {self.copies}"

    def get_short_info(self) -> str:
        return f"'{self.title}' ({self.author})"

    def update_copies(self, change: int) -> bool:
        new_copies = self.copies + change
        if new_copies < 0:
            print(f"Недостаточно экземпляров книги '{self.title}'. Доступно: {self.copies}")
            return False

        self.copies = new_copies
        action = "добавлено" if change > 0 else "уменьшено"
        print(f"Количество экземпляров книги '{self.title}' {action} на {abs(change)}. Теперь: {self.copies}")
        return True


class User:
    def __init__(self, name: str, card_number: str):
        self.name = name
        self.card_number = card_number
        self.borrowed_books: List[Book] = []
        print(f"Пользователь {name} (билет №{card_number}) зарегистрирован")

    def register(self) -> str:
        return f"Пользователь {self.name} (билет №{self.card_number}) успешно зарегистрирован"

    def add_borrowed_book(self, book: Book) -> bool:

        if book in self.borrowed_books:
            print(f"Книга '{book.title}' уже есть у пользователя {self.name}")
            return False

        self.borrowed_books.append(book)
        print(f"Книга '{book.title}' добавлена в список взятых пользователя {self.name}")
        return True

    def remove_borrowed_book(self, book: Book) -> bool:
        if book not in self.borrowed_books:
            print(f"Книга '{book.title}' не найдена в списке взятых пользователя {self.name}")
            return False

        self.borrowed_books.remove(book)
        print(f"Книга '{book.title}' удалена из списка взятых пользователя {self.name}")
        return True

    def get_borrowed_books_info(self) -> str:
        if not self.borrowed_books:
            return f"У пользователя {self.name} нет взятых книг"

        books_list = "\n  ".join([f"• {book.get_short_info()}" for book in self.borrowed_books])
        return f"Пользователь {self.name} взял следующие книги:\n  {books_list}"

    def get_user_info(self) -> str:
        return f"{self.name} | Билет №{self.card_number} | Взято книг: {len(self.borrowed_books)}"


class Library:
    def __init__(self, name: str = "Городская библиотека"):
        self.name = name
        self.books: List[Book] = []
        self.users: List[User] = []
        self.transactions: List[Dict[str, Union[str, datetime, Book, User]]] = []
        print(f"Библиотека '{name}' создана")

    def add_book(self, book: Book) -> None:
        existing_book = self.find_book_by_isbn(book.isbn)
        if existing_book:
            existing_book.update_copies(book.copies)
        else:
            self.books.append(book)

    def remove_book(self, book: Book) -> bool:
        if book not in self.books:
            print(f"Книга '{book.title}' не найдена в библиотеке")
            return False

        for user in self.users:
            if book in user.borrowed_books:
                print(f"Невозможно удалить книгу '{book.title}' — она выдана пользователю {user.name}")
                return False

        self.books.remove(book)
        print(f"Книга '{book.title}' удалена из библиотеки")
        return True

    def register_user(self, user: User) -> None:
        existing_user = self.find_user_by_card(user.card_number)
        if existing_user:
            print(f"Пользователь с билетом №{user.card_number} уже зарегистрирован")
        else:
            self.users.append(user)

    def deregister_user(self, user: User) -> bool:
        if user not in self.users:
            print(f"Пользователь {user.name} не найден в библиотеке")
            return False

        if user.borrowed_books:
            print(f"Невозможно удалить пользователя {user.name} — у него есть невозвращенные книги")
            return False

        self.users.remove(user)
        print(f"Пользователь {user.name} удален из библиотеки")
        return True

    def find_books_by_title(self, title: str) -> List[Book]:
        title_lower = title.lower()
        result = [book for book in self.books if title_lower in book.title.lower()]
        return result

    def find_books_by_author(self, author: str) -> List[Book]:
        author_lower = author.lower()
        result = [book for book in self.books if author_lower in book.author.lower()]
        return result

    def find_book_by_isbn(self, isbn: str) -> Optional[Book]:
        for book in self.books:
            if book.isbn == isbn:
                return book
        return None

    def find_user_by_name(self, name: str) -> List[User]:
        name_lower = name.lower()
        result = [user for user in self.users if name_lower in user.name.lower()]
        return result

    def find_user_by_card(self, card_number: str) -> Optional[User]:
        for user in self.users:
            if user.card_number == card_number:
                return user
        return None

    def issue_book(self, user: User, book: Book) -> bool:
        if book not in self.books:
            print(f"Книга '{book.title}' не найдена в библиотеке")
            return False

        if book.copies <= 0:
            print(f"Нет свободных экземпляров книги '{book.title}'")
            return False

        if user not in self.users:
            print(f"Пользователь {user.name} не зарегистрирован в библиотеке")
            return False

        if book in user.borrowed_books:
            print(f"Пользователь {user.name} уже взял книгу '{book.title}'")
            return False

        book.update_copies(-1)
        user.add_borrowed_book(book)

        transaction = {
            'type': 'issue',
            'user': user,
            'book': book,
            'date': datetime.now()
        }
        self.transactions.append(transaction)
        print(f"Книга '{book.title}' выдана пользователю {user.name}")
        return True

    def return_book(self, user: User, book: Book) -> bool:
        if book not in user.borrowed_books:
            print(f"Книга '{book.title}' не найдена в списке взятых пользователя {user.name}")
            return False

        book.update_copies(1)
        user.remove_borrowed_book(book)

        transaction = {
            'type': 'return',
            'user': user,
            'book': book,
            'date': datetime.now()
        }
        self.transactions.append(transaction)
        print(f"Книга '{book.title}' возвращена пользователем {user.name}")
        return True

    def display_transaction_history(self) -> None:
        if not self.transactions:
            print("История транзакций пуста")
            return

        print("\n" + "=" * 50)
        print("ИСТОРИЯ ТРАНЗАКЦИЙ")
        print("=" * 50)

        for i, transaction in enumerate(self.transactions, 1):
            date_str = transaction['date'].strftime("%Y-%m-%d %H:%M:%S")
            action = "ВЫДАЧА" if transaction['type'] == 'issue' else "ВОЗВРАТ"
            user_name = transaction['user'].name
            book_title = transaction['book'].title

            print(f"{i}. [{date_str}] {action}: '{book_title}' — {user_name}")

        print("=" * 50 + "\n")

    def display_all_books(self) -> None:
        if not self.books:
            print("В библиотеке нет книг")
            return

        print("\n" + "-" * 50)
        print(f"КНИГИ В БИБЛИОТЕКЕ '{self.name}'")
        print("-" * 50)
        for i, book in enumerate(self.books, 1):
            print(f"{i}. {book.get_info()}")
        print("-" * 50 + "\n")

    def display_all_users(self) -> None:
        if not self.users:
            print("В библиотеке нет зарегистрированных пользователей")
            return

        print("\n" + "-" * 50)
        print(f"ПОЛЬЗОВАТЕЛИ БИБЛИОТЕКИ '{self.name}'")
        print("-" * 50)
        for i, user in enumerate(self.users, 1):
            print(f"{i}. {user.get_user_info()}")
        print("-" * 50 + "\n")


def main():
    library = Library("Центральная городская библиотека")

    print("\n Создание книг")
    book1 = Book("Трудно быть Богом", "Бог", 0, "978-5-17-123456-7", 3)
    book2 = Book("Руководство по эксплоатации танка Т-44", "А.М. Катунский", 1946, "978-5-04-123789-4", 2)
    book3 = Book("Охота на крокодила со спинингом на средней Волге", "М.Н. Никольской", 1768, "978-5-699-12345-6", 2)
    book4 = Book("Искусство рассуждать о книгах, которых вы не читали", "Пьер Байяр", 2006, "978-5-04-456789-1", 1)
    book5 = Book("Дневник 3", "Стенфорд Пайнс", 1982, "978-5-17-789123-4", 5)

    print("\n Добавление книг в библиотеку")
    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)
    library.add_book(book4)
    library.add_book(book5)

    print("\n Регистрация пользователей")
    user1 = User("Лиза Барбоскина", "LIB-001")
    user2 = User("Дядя Шнюк", "LIB-002")
    user3 = User("Дедус", "LIB-003")

    library.register_user(user1)
    library.register_user(user2)
    library.register_user(user3)

    library.display_all_books()
    library.display_all_users()

    print("\n Выдача книг")
    library.issue_book(user1, book1)
    library.issue_book(user1, book2)
    library.issue_book(user2, book1)
    library.issue_book(user3, book3)
    library.issue_book(user3, book5)
    library.issue_book(user2, book4)
    library.issue_book(user2, book4)

    print("\n Состояние после выдачи")
    library.display_all_books()
    print(user1.get_borrowed_books_info())
    print(user2.get_borrowed_books_info())
    print(user3.get_borrowed_books_info())

    print("\n Возврат книг")
    library.return_book(user1, book1)
    library.return_book(user2, book4)

    print("\n Выдыча после возврата")
    library.issue_book(user2, book1)

    print("\n Поиск книг")
    god_books = library.find_books_by_author("Бог")
    print(f"Книги автора 'Бог':")
    for book in god_books:
        print(f"  • {book.get_short_info()}")

    fishing_books = library.find_books_by_title("крокодил")
    print(f"\nКниги со словом 'крокодил' в названии:")
    for book in fishing_books:
        print(f"  • {book.get_short_info()}")

    print("\n Поиск пользователей")
    barboskin_users = library.find_user_by_name("Барбоскина")
    print(f"Пользователи с фамилией 'Барбоскина':")
    for user in barboskin_users:
        print(f"  • {user.get_user_info()}")

    library.display_transaction_history()



if __name__ == "__main__":
    main()
````
