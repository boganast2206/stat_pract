## Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt


def beauty_derivative_plot(x0):
    x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
    f = 2 * np.sin(x) - np.cos(2 * x)
    df = 2 * np.cos(x) + 2 * np.sin(2 * x)
    f0 = 2 * np.sin(x0) - np.cos(2 * x0)
    df0 = 2 * np.cos(x0) + 2 * np.sin(2 * x0)
    tangent = f0 + df0 * (x - x0)

    plt.figure(figsize=(10, 5))

    plt.plot(x, f, color='blue',
             label='f(x) = 2sin x âˆ’ cos 2x')
    plt.plot(x, df, '--', color='red',
             label="f'(x) = 2cos x + 2sin 2x")
    plt.plot(x, tangent, color='green',
             linewidth=2,
             label=f'Tangent at x = {x0:.2f}')

    plt.scatter(x0, f0, color='black', zorder=5)

    plt.plot([x0, x0], [0, f0], linestyle='--', color='purple')

    if x0 in [-3.0, -2.0]:
        plt.plot([x0, x0 + 0.8], [f0, f0], linestyle='--', color='purple')
    elif x0 == 1.7:
        plt.plot([x0, x0 - 0.8], [f0, f0], linestyle='--', color='purple')
    else:
        if df0 > 0:
            plt.plot([x0, x0 + 0.8], [f0, f0], linestyle='--', color='purple')
        else:
            plt.plot([x0, x0 - 0.8], [f0, f0], linestyle='--', color='purple')

    plt.annotate(
        f"f'({x0:.2f}) = {df0:.2f}",
        xy=(x0, f0),
        xytext=(x0 - 0.7, f0 - 1.4),
        arrowprops=dict(arrowstyle='->')
    )

    plt.title('Function, Derivative, and Tangent Line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-4, 4)
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()


beauty_derivative_plot(-3.0)
beauty_derivative_plot(-2.0)
beauty_derivative_plot(1.7)
```
